#  Análise de Dados

Uma simples análise dos dados de treino e teste.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Diretórios 
source_path = '../data/raw/'
data_path = '../data/processed/'
```

```python
# Carregando os dados
train_images = np.load(data_path+"train_images.npy")
train_labels = np.load(data_path+"train_labels.npy")
test_images = np.load(data_path+"test_images.npy")
test_labels = np.load(data_path+"test_labels.npy")
```

```python
# Estrutura dos dados
print("train_images:",train_images.shape)
print("train_labels:",train_labels.shape)
print("test_images:",test_images.shape)
print("test_labels:",test_labels.shape)
```

```python
# Amostra
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
# Normalizando
train_images = train_images / 255.0

test_images = test_images / 255.0
```

```python
# Resultado
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
def plot_quantity_by_class(df):
    df.columns = ['labels']
    ctlabels=df.groupby('labels')['labels'].count()
    fig, ax = plt.subplots(figsize=(10,6))
    plt.xticks(ctlabels.index)
    plot = ax.bar(ctlabels.index,ctlabels.values)
    for rect in plot:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.002*height,'%d' %
        int(height), ha='center', va='bottom')
```

```python
# Balanço Dataset Treino
df_train = pd.DataFrame(train_labels)
plot_quantity_by_class(df_train)
```

```python
# Balanço Dataset Teste
df_test = pd.DataFrame(test_labels)
plot_quantity_by_class(df_test)
```

```python
# Amostras
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(train_labels[i])
plt.show()
```
